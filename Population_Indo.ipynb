```python
import numpy as np
import pandas as pd
import gdown
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_url = 'https://docs.google.com/spreadsheets/d/148yDWEfsbZemusKwsRCcdc3ECu7hq79NeEsinbAjuP4/edit?usp=sharing'
output_file = 'DataPopulasi.xlsx'
gdown.download(file_url, output_file, quiet=False)
df = pd.read_excel(output_file)
print(df)
```

```python
df
```

```python
#Percipitation (rain) Meters SDMX code**
df.rename(columns={'Region, subregion, country or area *': 'Area'}, inplace=True)
df.rename(columns={'Location code': 'LocationCode'}, inplace=True)
df.rename(columns={'ISO3 Alpha-code': 'IS03code'}, inplace=True)
df.rename(columns={'ISO2 Alpha-code': 'IS02code'}, inplace=True)
df.rename(columns={'Male Population, as of 1 July (thousands)': 'male'}, inplace=True)
df.rename(columns={'Female Population, as of 1 July (thousands)': 'female'}, inplace=True)
df.rename(columns={'Population Density, as of 1 July (persons per square km)': 'density'}, inplace=True)
df.rename(columns={'Population Growth Rate (percentage)': 'growth'}, inplace=True)
df.rename(columns={'SDMX code**': 'SDMX'}, inplace=True)
```

```python
# Filter rows where SDMX_code is 1 (World)
df_world = df[df['SDMX'] == 1]

# Filter rows where SDMX_code is 360 (Indonesia)
df_indonesia = df[df['SDMX'] == 360]
```

```python
# Convert total population column to string
df['Total Population, as of 1 January (thousands)'] = df['Total Population, as of 1 January (thousands)'].astype(str)

# Remove decimal points from thousands separator
df['Total Population, as of 1 January (thousands)'] = df['Total Population, as of 1 January (thousands)'].str.replace('.', '')

# Convert total population column to numeric data type
df['Total Population, as of 1 January (thousands)'] = pd.to_numeric(df['Total Population, as of 1 January (thousands)'], errors='coerce')

# Filter data for the year 2000 onwards
start_year = 2000
df_world = df_world[df_world['Year'] >= start_year]
df_indonesia = df_indonesia[df_indonesia['Year'] >= start_year]

# Calculate annual population growth
df_world['Population_Growth_Rate'] = df_world['Total Population, as of 1 January (thousands)'].pct_change() * 100
df_indonesia['Population_Growth_Rate'] = df_indonesia['Total Population, as of 1 January (thousands)'].pct_change() * 100

# Visualize line graph
plt.figure(figsize=(10, 6))
plt.plot(df_world['Year'], df_world['Population_Growth_Rate'], label='World')
plt.plot(df_indonesia['Year'], df_indonesia['Population_Growth_Rate'], label='Indonesia')

plt.title('Population Growth: World vs Indonesia (2000 - Present)')
plt.xlabel('Year')
plt.ylabel('Population Growth (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter data untuk tahun 2000 hingga saat ini
start_year = 2000
df_world_growth = df_world[df_world['Year'] >= start_year]
df_indonesia_growth = df_indonesia[df_indonesia['Year'] >= start_year]

# Hitung total pertumbuhan populasi dunia untuk setiap tahun
total_growth_world = df_world_growth.groupby('Year')['Population Change (thousands)'].sum()

# Hitung total pertumbuhan populasi Indonesia untuk setiap tahun
total_growth_indonesia = df_indonesia_growth.groupby('Year')['Population Change (thousands)'].sum()

# Hitung persentase kontribusi Indonesia terhadap pertumbuhan populasi dunia untuk setiap tahun
percentage_contribution = (total_growth_indonesia / total_growth_world) * 100

# Visualisasi diagram lingkaran
plt.figure(figsize=(8, 8))
plt.pie(percentage_contribution, labels=percentage_contribution.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Percentage Contribution of Indonesia to World Population Growth (2000 - 2021)', fontsize=16)
plt.axis('equal')  # Membuat lingkaran terlihat sempurna
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Filter data untuk Indonesia
df_indonesia_growth = df_indonesia.groupby('Year')['Population Change (thousands)'].sum().reset_index()

# Persiapkan data untuk regresi linier
X = df_indonesia_growth['Year'].values.reshape(-1, 1)
y = df_indonesia_growth['Population Change (thousands)'].values

# Membuat model regresi linier
model = LinearRegression()
model.fit(X, y)

# Membuat proyeksi pertumbuhan populasi Indonesia di masa depan (misalnya, untuk 10 tahun ke depan)
future_years = np.arange(df_indonesia_growth['Year'].max() + 1, df_indonesia_growth['Year'].max() + 11).reshape(-1, 1)
future_growth = model.predict(future_years)

# Hitung proyeksi dampaknya terhadap populasi dunia
world_impact_percentage = (future_growth.sum() / df_indonesia_growth['Population Change (thousands)'].sum()) * 100

# Visualisasi grafik garis
plt.figure(figsize=(10, 6))

# Plot data historis pertumbuhan populasi Indonesia
plt.plot(df_indonesia_growth['Year'], df_indonesia_growth['Population Change (thousands)'], marker='o', label='Historical Growth (Indonesia)')

# Plot proyeksi pertumbuhan populasi Indonesia di masa depan
plt.plot(np.concatenate([df_indonesia_growth['Year'], future_years.flatten()]), np.concatenate([df_indonesia_growth['Population Change (thousands)'], future_growth]), linestyle='--', color='orange', label='Projected Growth (Indonesia)')

# Label dan judul grafik
plt.title('Population Growth Projection: Indonesia and Impact on World Population', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population Growth (thousands)', fontsize=14)
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

# Tampilkan dampaknya terhadap populasi dunia
print(f"Projected Impact on World Population: {world_impact_percentage:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_indonesia and df_world are already defined with columns: year, male, female

# Filter the data for the year 2021
indonesia_2024 = df_indonesia[df_indonesia['Year'] == 2021]
world_2024 = df_world[df_world['Year'] == 2021]

# Check if data for 2024 is available
if indonesia_2024.empty or world_2024.empty:
    print("No data available for the year 2021.")
else:
    # Calculate the total male population for Indonesia and the World in 2024
    total_indonesia_male_2024 = indonesia_2024['male'].sum()
    total_world_male_2024 = world_2024['male'].sum()

    # Calculate the percentage of Indonesian males out of the total world males in 2024
    percent_indonesian_male = (total_indonesia_male_2024 / total_world_male_2024) * 100

    # Pie chart data
    labels = ['Indonesian Male', 'Other World Male']
    sizes = [percent_indonesian_male, 100 - percent_indonesian_male]
    colors = ['lightcoral', 'lightskyblue']

    # Plotting the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    plt.title('Percentage of Indonesian Male Population in 2021 compared to World Male Population')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```
